```python
import os
from pathlib import Path
import pandas as pd
```

```python
csv_file_path = Path().cwd().parent / 'data' / 'ResalePricesSingapore.csv'
df = pd.read_csv(csv_file_path)
df['month'] = pd.to_datetime(df['month'])

display(df.head(5), df.info(), df.describe())
```

```python
# Get subset based on last 3 digits of matriculation number
# 346 -> 6: YYYY = 2016; 4: MM = 04; 3: town = CHOA CHU KANG
yyyy = 2016
mm = 4
town = 'CHOA CHU KANG'

df = df.loc[(df['month'] >= pd.to_datetime(f'{yyyy}-{mm}')) & (df['month'] <= pd.to_datetime(f'{yyyy}-{mm+1}')) & (df['town']==town), :]
```

```python
# example queries
# Minimum Price, Standard Deviation of Price, Average Price, Minimum Price per Square Meter

print(f"Minimum Price = {df['resale_price'].min()}")
print(f"Standard Deviation of Price = {df['resale_price'].std()}")
print(f"Average Price = {df['resale_price'].mean()}")
print(f"Minimum Price per Square Meter = {(df['resale_price']/df['floor_area_sqm']).min()}")
```
